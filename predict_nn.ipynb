```python
import pandas as pd
# æ•°æ®åŠ è½½
train_data = pd.read_csv('./used_car_train_20200313.csv',sep = ' ')
train_data
test = pd.read_csv('./used_car_testB_20200421.csv',sep=' ')
test
```

```python
train_data.replace('-','0.0',inplace = True)
test.replace('-','0.0',inplace = True)
```

```python
# train_data['power'].describe()
train_data['power'][train_data['power'] >600] = 600
train_data['power'].describe()
test['power'][test['power'] >600] = 600
test['power'].describe()
```

```python
feature_cols = train_data.columns.tolist()
feature_cols.remove('price')
print(feature_cols)
```

```python
train_data.fillna(train_data.median(), inplace= True)
test.fillna(test.median(), inplace= True)
```

```python
# æ•°æ®å½’ä¸€åŒ–å¤„ç†
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
x = mm.fit_transform(train_data[feature_cols])
y = train_data['price'].values
```

```python
# åˆ‡åˆ†æ•°æ®é›†
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2, random_state = 2021)
```

```python
from tensorflow import keras
# æ­å»ºæ¨¡åž‹
model = keras.Sequential([
    # 250ç»´åº¦ å®šä¹‰å‘é‡çš„é•¿åº¦ å»ºè®®åœ¨200-300 
    keras.layers.Dense(320, activation='relu', input_shape = [len(feature_cols)]),
    keras.layers.Dense(160, activation='relu'),
    keras.layers.Dense(80, activation='relu'),
    # æœ€åŽçš„è¾“å‡ºç»“æžœæ˜¯price
    keras.layers.Dense(1)
    
])
# è®¾ç½®ä¼˜åŒ–å™¨
model.compile(loss='mean_absolute_error',optimizer = 'Adam')
model.fit(train_x, train_y, batch_size = 2048, epochs = 100)
```
